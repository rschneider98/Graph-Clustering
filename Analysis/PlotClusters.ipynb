```python
import numpy as np
import networkx as nx
from scipy.io import mmread
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

# Get Eigenvector Calculations and Original Graph

```python
raw_input_matrix = mmread('../Data/testEigenvectors.mtx')
```

```python
eigenvectors = raw_input_matrix.todense()
```

```python
eigenvectors.shape
```

```python
raw_input_matrix = mmread('../Data/test.mtx')
```

```python
G = nx.Graph(raw_input_matrix.todense())
```

# Cluster the Eigenvectors

```python
kmeans = KMeans(init="k-means++", n_clusters=5, n_init=10,
                random_state=0)
clusters = kmeans.fit(eigenvectors)
```

```python
clusters.labels_
```

```python
cmap = plt.get_cmap("tab10")
```

```python
colors = [cmap(x) for x in clusters.labels_.tolist()]
```

```python
nx.draw(G, node_color=colors, node_size=100, linewidths=0.15)
```

```python
kmeans = KMeans(init="k-means++", n_clusters=5, n_init=10,
                random_state=0)
clusters = kmeans.fit(eigenvectors[:,6:9]) # use the eigenvectors of the three larges eigenvalues for clustering
```

```python
clusters.labels_
```

```python
colors = [cmap(x) for x in clusters.labels_.tolist()]
```

```python
nx.draw(G, node_color=colors, node_size=100, linewidths=0.15)
```

```python
nx.draw_shell(G, node_color=colors, node_size=100, linewidths=0.15)
```
