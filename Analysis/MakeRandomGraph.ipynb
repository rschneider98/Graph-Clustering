```python
import numpy as np
import networkx as nx
from scipy import sparse
from scipy.io import mmwrite
import matplotlib.pyplot as plt
```

```python
number_of_groups = 5
nodes_per_group = 10
```

```python
number_of_groups * nodes_per_group
```

```python
G = nx.connected_caveman_graph(number_of_groups, nodes_per_group)
```

```python
nx.draw(G)
```

```python
nx.draw_random(G)
```

```python
nx.draw_spectral(G)
```

```python
A = nx.to_numpy_matrix(G)
```

```python
L = nx.linalg.laplacianmatrix.laplacian_matrix(G)
```

```python
A_out = sparse.csr_matrix(A)
L_out = sparse.csr_matrix(L)
```

```python
mmwrite("../Data/test.mtx", A_out, precision=6)
mmwrite("../Data/testLaplacian.mtx", L_out, precision=6)
```
