```python
import numpy as np
import networkx as nx
from scipy.io import mmread
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

# Read in the input matrix of Power Network
http://networkrepository.com/power-494-bus.php

```python
raw_input_matrix = mmread('../Data/power-494-bus.mtx')
```

```python
input_matrix = raw_input_matrix.todense()
```

```python
input_matrix.shape
```

# Find Eigenvalues and Eigenvectors

```python
eigenvalues, eigenvectors = np.linalg.eig(input_matrix)
```

```python
print(eigenvalues)
```

# Basic Plot of Data

```python
G = nx.Graph(input_matrix)
nx.draw(G, node_size=10, linewidths=0.15)
```

```python
nx.draw_spectral(G, node_size=10)
```

# Get first two eigenvectors and use as (x, y) coordiantes

```python
coord = eigenvectors[:, :2].tolist()
pos = {}
for i in range(len(coord)):
    pos.update({i: (coord[i][0], 0)})
nx.draw(G, pos)
```

```python
coord = eigenvectors[:, :2].tolist()
pos = {}
for i in range(len(coord)):
    pos.update({i: (coord[i][0], coord[i][1])})

nx.draw(G, pos)
```

# KMeans Clustering

```python
kmeans = KMeans(init="k-means++", n_clusters=10, n_init=4,
                random_state=0)
clusters = kmeans.fit(eigenvectors)
```

```python
clusters.labels_
```

```python
cmap = plt.get_cmap("tab10")
```

```python
colors = [cmap(x) for x in clusters.labels_.tolist()]
```

```python
nx.draw_shell(G, node_color=colors, node_size=10, linewidths=0.15)
```

```python
nodes = []
colors = []
for i in range(len(clusters.labels_)):
    if clusters.labels_[i] == 0:
        nodes.append(i)
        colors.append(cmap(clusters.labels_[i]))

nx.draw(G, nodelist=nodes, node_color=colors, node_size=100, linewidths=0.15)
```

```python
nx.draw(G.subgraph(nodes), node_color=colors, node_size=100, linewidths=0.15)
```
